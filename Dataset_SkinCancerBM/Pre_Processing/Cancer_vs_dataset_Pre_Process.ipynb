## Extract Raw Dataset

```python
import zipfile

with zipfile.ZipFile("/content/drive/MyDrive/Student/Mohit/Dataset/Skin_Cancer_Malignant_vs_Benign/archive.zip", 'r') as zip_ref:
    zip_ref.extractall("/content")
```

```python
import os
import pandas as pd

# Function to create dataframe from a directory
def create_dataframe(data_dir):
    data = []
    for label in ['benign', 'malignant']:
        class_dir = os.path.join(data_dir, label)
        if os.path.exists(class_dir):
            for filename in os.listdir(class_dir):
                file_path = os.path.join(class_dir, filename)
                data.append([file_path, label])

    df = pd.DataFrame(data, columns=['file_path', 'label'])
    return df

train_dir = '/content/data/train'
test_dir = '/content/data/test'

train_df = create_dataframe(train_dir)
_test_df = create_dataframe(test_dir)

print("Train DataFrame:")
train_df
```

```python
print("Test DataFrame:")
_test_df
```

**Now spliting test set into test and validation by 50% and saving these dataframe as const to drive.**

```python
from sklearn.model_selection import train_test_split

# Split test_df into test and validation sets (50/50 split)
test_df, val_df = train_test_split(_test_df, test_size=0.5, random_state=42)

train_save_path = '/content/drive/MyDrive/Student/Mohit/Dataset/Federated_Learning_Project_2/Processed_CSV/train_df.csv'
test_save_path = '/content/drive/MyDrive/Student/Mohit/Dataset/Federated_Learning_Project_2/Processed_CSV/test_df.csv'
val_save_path = '/content/drive/MyDrive/Student/Mohit/Dataset/Federated_Learning_Project_2/Processed_CSV/val_df.csv'

train_df.to_csv(train_save_path, index=False)
test_df.to_csv(test_save_path, index=False)
val_df.to_csv(val_save_path, index=False)

print(f"Train dataframe saved to: {train_save_path}")
print(f"Test dataframe saved to: {test_save_path}")
print(f"Validation dataframe saved to: {val_save_path}")
```

## A Small Cross Checking that all goes well by ploting images from each dataframe

```python
import matplotlib.pyplot as plt
from PIL import Image

def display_images(dataframe, title):
    print(f"\n{title}:")
    plt.figure(figsize=(10, 5))
    for i in range(min(3, len(dataframe))):
        img_path = dataframe.iloc[i]['file_path']
        label = dataframe.iloc[i]['label']
        try:
            img = Image.open(img_path)
            plt.subplot(1, 3, i + 1)
            plt.imshow(img)
            plt.title(f"Label: {label}")
            plt.axis('off')
        except FileNotFoundError:
            print(f"Error: Image file not found at {img_path}")
        except Exception as e:
            print(f"Error loading image {img_path}: {e}")
    plt.show()

display_images(train_df, "First 3 Images from Train DataFrame")
display_images(test_df, "First 3 Images from Test DataFrame")
display_images(val_df, "First 3 Images from Validation DataFrame")
```
